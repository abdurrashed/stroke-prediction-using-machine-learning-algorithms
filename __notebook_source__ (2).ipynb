```none
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```none
df= pd.read_csv('/content/drive/MyDrive/dmmml project/healthcare-dataset-stroke-data.csv')
```

```none
df.head()
```

```none
df.shape
```

```none
# Dataset info
df.info()
```

```none
# Check missing value
df.isnull().sum()
```

```none
# Drop row with missing value
df = df.dropna(axis=0)
df
```

```none
# Drop id 
df = df.drop(['id'], axis=1)
df
```

```none
from sklearn.preprocessing import LabelEncoder
from pandas.core.dtypes.common import is_numeric_dtype

# Label Encoding
le = LabelEncoder() # method initialization

for col in df.columns :
   if is_numeric_dtype(df[col]):
       continue

   df[col]=le.fit_transform(df[col])

df.head()
```

```none
# Define X (Feature) and Y (Target)


X = df.drop('stroke',axis=1)

y = df['stroke']
```

```none
# Split dataset to traning and testing set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```none
# Modelling
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC


classifier_KNN = KNeighborsClassifier()
classifier_DT = DecisionTreeClassifier()
classifier_GNB = GaussianNB()
classifier_log=LogisticRegression()
```

### **1.KNN**

```none
classifier_KNN.fit(X_train, y_train)
```

```none
KNN = classifier_KNN.predict(X_test)
dataEvaluation=pd.DataFrame({"y_Predict":KNN, "y_Actual":y_test})
dataEvaluation
```

```none
classifier_KNN.score(X_test,y_test)
```

```none
y_prediction = classifier_KNN.predict(X_test)
y_prediction
```

```none
from sklearn.metrics import  confusion_matrix
cm = confusion_matrix(y_test, y_prediction)
cm
```
